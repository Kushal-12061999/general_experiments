```python
from transformers import AutoTokenizer, AutoModelForCausalLM
import torch
```

```python
# Pick any SLM on HuggingFace
MODEL_NAME = "microsoft/phi-3-mini-4k-instruct"
# Other options:
# "microsoft/phi-2"
# "meta-llama/Llama-3.2-1B-Instruct"
# "TinyLlama/TinyLlama-1.1B-Chat-v1.0"

device = "cuda"  # FORCE GPU

tokenizer = AutoTokenizer.from_pretrained(MODEL_NAME)

model = AutoModelForCausalLM.from_pretrained(
    MODEL_NAME,
    torch_dtype=torch.float16,   # GPU-friendly
    device_map={"": "cuda"}      # FORCE GPU
).eval()
```

```python
def prompt_generator_summary(call_transcript):
    prompt = """
<|system|>
You are an AI assistant that summarizes customer service call transcripts.
Ignore the virtual assistant instructions like “press a number for an option.”
Focus only on the conversation between human participants.
Return ONLY the summary inside <summary>...</summary>.
Do NOT add anything else.

<|user|>
Here is the transcript:

<transript>
{transcript}
</transcript>

Please provide the summary.

<|assistant|>""".format(transcript = call_transcript)
    return prompt
```

```python
call_example = """Agent: Hey, Good Morning!
Customer: Hello, this is Kushal.
Agent: Tell me, how can I help you today?
Customer: I’m having some issues with my SIM card network.
Agent: I can help you with that. Since when have you been facing this issue?
Customer: It started this morning, suddenly there’s no signal.
Agent: Alright, could you please confirm your mobile number?
Customer: Sure, it’s 9823XXXX45.
Agent: Thank you. Please try restarting your phone once; this usually fixes network drops.
Customer: I just restarted, and yes, the network is back. Thanks a lot!"""
```

```python
prompt = prompt_generator_summary(call_example)
```

```python
print(prompt)
```

```python
inputs = tokenizer(prompt, return_tensors="pt").to(device)

with torch.no_grad():
    outputs = model.generate(
        **inputs,
        max_new_tokens=50,
        temperature=0.1,
    )

print(tokenizer.decode(outputs[0], skip_special_tokens=True))
```
